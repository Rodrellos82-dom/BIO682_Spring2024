```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
```

# Read in data

```python
infile = "/Users/jtladner/Documents/GitHub/BIO682_Spring2023/Week10_Matplotlib-pt1/Assignment/importexport.txt"

exD = defaultdict(list)
imD = defaultdict(list)

with open(infile, "r") as fin:
    next(fin)
    for line in fin:
        cols = line.rstrip("\n").split("\t")
        exD[cols[0]].append(float(cols[4]))
        imD[cols[0]].append(float(cols[3]))
        
```

# Generate plot

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(imD["WesternUrban"], exD["WesternUrban"], label="WesternUrban", c="green")
ax.scatter(imD["WesternRural"], exD["WesternRural"], label="WesternRural", c="blue")
ax.set_xlabel("Number of Imports", fontsize=20)
ax.set_ylabel("Number of Exports", fontsize=20)
ax.legend()
```

# Read in whole data file using Pandas

```python
df = pd.read_csv(infile, sep="\t", header=0)
```

```python
wu = df[df["District"]=="WesternUrban"]
wr = df[df["District"]=="WesternRural"]
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(wu["Imports"], wu["Exports"], label="WesternUrban", c="green")
ax.scatter(wr["Imports"], wr["Exports"], label="WesternRural", c="blue")
ax.set_xlabel("Number of Imports", fontsize=20)
ax.set_ylabel("Number of Exports", fontsize=20)
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.hist(wu["Imports"], label="WesternUrban", color="orange", bins=10, alpha=0.7, zorder=2)
ax.hist(wr["Imports"], label="WesternRural", color="blue", bins=20, alpha=0.7, zorder=1)
ax.set_xlabel("Number of Imports", fontsize=20)
ax.set_ylabel("Frequency", fontsize=20)
ax.legend()
```
