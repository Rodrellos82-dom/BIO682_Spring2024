```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csF = "/Users/jtladner/Documents/GitHub/BIO682_Spring2024/Week10_Matplotlib-pt1/Demo/normReadCounts.tsv"
csDF = pd.read_csv(csF, sep="\t", header=0, index_col=0)
```

```python
enrF = "/Users/jtladner/Documents/GitHub/BIO682_Spring2024/Week10_Matplotlib-pt1/Demo/enrichedPeptides.txt"
enrPeps = pd.read_csv(enrF, header=None, names=["PepName"])
```

```python
yNames = [n for n in csDF.columns if n.startswith("ZN-168_pG")]
xNames = [n for n in csDF.columns if n.startswith("SB")]

yDF = csDF[yNames]
xDF = csDF[xNames]

y = yDF.mean(axis=1)
x = xDF.mean(axis=1)

enrY = y.loc[enrPeps["PepName"]]
enrX = x.loc[enrPeps["PepName"]]
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(np.log10(x+1), np.log10(y+1), c="grey", alpha=0.2, s=6)
ax.scatter(np.log10(enrX+1), np.log10(enrY+1), c="orange", alpha=0.8, s=20)
ax.set_xlabel("Negative Control", fontsize=20)
ax.set_ylabel("Bat Sample", fontsize=20)
ax.text(1.62, 2.5, "Enriched", c="orange", fontsize=20, ha="right", va="center")

fig.savefig("Demo2.pdf", bbox_inches="tight", dpi=300)
fig.savefig("Demo2.png", bbox_inches="tight", dpi=300)
```
